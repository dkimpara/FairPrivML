Preprocessing for Diabetes data.

```python
import pandas as pd
```

```python
data = pd.read_csv("diabetic_data.csv")
```

```python
ageReplace = {'[0-10)' : 5, '[10-20)' : 15, '[20-30)' : 25,
               '[30-40)' : 35, '[40-50)' : 45,
               '[50-60)' : 55, '[60-70)' : 65,
               '[70-80)' : 75, '[80-90)' : 85,
               '[90-100)' : 95}
```

```python
age = data['age']
age = age.replace(ageReplace)
data['age'] = age
data
```

```python
re = data["readmitted"]
re = re.replace({">30":1, "<30":1, "NO":0})
data['readmitted'] = re
data
```

```python
gender = data["gender"]
gender = gender.replace({"Male":0, "Female":1, "Unknown/Invalid":1})
data['gender'] = gender
data
```

```python
data.groupby(["gender", "readmitted"]).size()
```

```python
diabetic_females = data[(data["gender"]==1) & (data["readmitted"]==1)]
diabetic_females = diabetic_females.copy()
```

```python
indices = list(diabetic_females.index.values)
indices
print(len(indices))
half = indices[:int(len(indices)/2)]
print(len(half))
```

```python
len(diabetic_females.index)
diabetic_females.drop(labels=half, inplace=True)
len(diabetic_females.index)
```

```python
updated = pd.concat([data, diabetic_females]).drop_duplicates(keep=False)
len(updated.index)
```

```python
updated.to_csv("updated_data", sep=" ", index=False)
```
