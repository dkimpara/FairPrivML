Preprocessing for Diabetes data.

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("diabetes_data/diabetic_data.csv")
```

```python
ageReplace = {'[0-10)' : 5, '[10-20)' : 15, '[20-30)' : 25,
               '[30-40)' : 35, '[40-50)' : 45,
               '[50-60)' : 55, '[60-70)' : 65,
               '[70-80)' : 75, '[80-90)' : 85,
               '[90-100)' : 95}
```

```python
age = data['age']
age = age.replace(ageReplace)
data['age'] = age
data
```

```python
re = data["readmitted"]
re = re.replace({">30":1, "<30":1, "NO":0})
data['readmitted'] = re
data
```

```python
gender = data["gender"]
gender = gender.replace({"Male":0, "Female":1, "Unknown/Invalid":1})
data['gender'] = gender
data
```

```python
data.groupby(["gender", "readmitted"]).size()
```

```python
diabetic_females = data[(data["gender"]==1) & (data["readmitted"]==1)]
diabetic_females = diabetic_females.copy()
```

```python
indices = list(diabetic_females.index.values)
indices
print(len(indices))
half = indices[:int(len(indices)/2)]
print(len(half))
```

```python
len(diabetic_females.index)
diabetic_females.drop(labels=half, inplace=True)
len(diabetic_females.index)
```

```python
updated = pd.concat([data, diabetic_females]).drop_duplicates(keep=False)
len(updated.index)
```

```python
updated
```

```python
updated.to_csv("updated_data", sep=" ", index=False)
```

Reloading to drop appropriate columns, one hot, and binarize.

```python
import pandas as pd
import numpy as np
d = pd.read_csv("diabetes_data/updated_data", delim_whitespace=True)
d
```

```python
d['diabetesMed'] = d['diabetesMed'].replace({"No":0, "Yes":1})
d
```

```python
d['change'] = d['change'].replace({"No":0, "Ch":1})
```

```python
d.drop([], axis=1)
```

```python
notOneHot = ['age', 'time_in_hospital', 'num_lab_procedures',
            'num_procedures', 'num_medications','number_outpatient',
            'number_emergency', 'number_diagnoses', 'diabetesMed', 'readmitted', 'gender', 'change']
cols = d.columns.tolist()
for i in notOneHot:
    cols.remove(i)
cols
```

```python
dummies = pd.get_dummies(d, columns=cols)
```

```python
dummies
```

```python
columns = dummies.columns.tolist()
print len(columns)
columns.remove("gender")
print len(columns)
columns.remove("readmitted")
print len(columns)
columns.append("gender")
columns.append("readmitted")
```

```python
dummies = dummies[columns]
dummies
```

```python
dummies.info()
```

```python
dummies = dummies.sample(frac=1).reset_index(drop=True)
dummies
```

```python
train, test = np.split(dummies.sample(frac=1), [int(.6*len(dummies))])
print(len(train))
print(len(test))
print(len(train) + len(test))
```

```python
train.to_csv("updated-train", sep=" ", header=False, index=False)
```

```python
test.to_csv("updated-test", sep=" ", header=False, index=False)
```
